In the context of this project, where the primary objective is to build a **machine learning-based smart consensus model** that **outperforms the naive average** of analyst estimates, it is essential to ensure that the data used for training reflects the actual **information environment** analysts would have had at the time of making their forecasts. The key consideration is how to appropriately **align macroeconomic indicators** with each forecasted period in a way that **avoids lookahead bias** and maintains the integrity of the model.

The dataset contains analyst estimates with a target period specified by the variable **"Period"**, which refers to the **timeframe being forecasted** (e.g., Q3 Sep-11), not the date when the forecast was issued. Because we do not observe the exact **timestamp of each estimate**, we must simulate a realistic information set that approximates what analysts could have known **prior to the forecasted period**. Using macroeconomic data from the same period, or worse, from the end of the period, would inject **future information** into the model and **invalidate its predictive power**.

To address this, we **map each forecasted period to macroeconomic data from one month before the beginning of that period**. This is based on a practical assumption: most analyst forecasts are made shortly before the target period begins, and macroeconomic statistics for any given month are typically **published early the following month**. By selecting data from the **last available day of the previous month**, we ensure that we are only using information that would have been **realistically available to analysts** at the time of their estimation.

This approach avoids **lookahead bias**, supports **generalization**, and ensures that the model is trained on features that were truly **observable at forecast time**. It works across all forecast horizons: **quarterly, half-yearly, and full-year**. For example, a forecast labeled “Q3 Sep-11” (referring to July–September 2011) would use macro data from **June 2011**; a full-year forecast labeled “FY Dec-11” would use data from **December 2010**; and a half-year forecast labeled “H2 Dec-17” would use data from **June 2017**. 

| **Forecast Horizon** | **Period Label** | **Interpreted Period** | **Use Macro Data From** |
|----------------------|------------------|-------------------------|--------------------------|
| Quarter              | Q3 Sep-11        | Jul–Sep 2011            | June 2011                |
| Quarter              | Q4 Dec-11        | Oct–Dec 2011            | September 2011           |
| FY                   | FY Dec-11        | Jan–Dec 2011            | December 2010            |
| H1                   | SAN1 Jun-18      | Jan–Jun 2018            | December 2017            |
| H2                   | SAN2 Dec-17      | Jul–Dec 2017            | June 2017                |

```python
macro_df["Date"] = pd.to_datetime(macro_df["Date"])
macro_df.head()
```

```python
from dateutil import parser
from pandas.tseries.offsets import MonthEnd

# We define a function to convert 'Period' + 'Horizon' into a datetime
def period_to_date(row):
    period = row["Period"]
    horizon = row["Horizon"]

    try:
        # General case: use last part of the string as month-year (e.g., "Sep-11", "Dec-12")
        if horizon in ["Quarter", "FY"]:
            date_str = period.split()[-1]
            dt = parser.parse("01-" + date_str, dayfirst=True) + MonthEnd(1)
        elif horizon == "H1":
            # For 'SAN1 Jun-20' → take the 'Jun-20'
            date_str = period.split()[-1]
            dt = parser.parse("01-" + date_str, dayfirst=True) + MonthEnd(1)
        elif horizon == "H2":
            # For 'SAN2 Dec-20' → same logic
            date_str = period.split()[-1]
            dt = parser.parse("01-" + date_str, dayfirst=True) + MonthEnd(1)
        else:
            return pd.NaT
    except Exception:
        return pd.NaT

    return dt

# We create 'Macro_Merge_Date' = 1 month before target period
real_df["Target_Period_End"] = real_df.apply(period_to_date, axis=1)
real_df["Macro_Merge_Date"] = real_df["Target_Period_End"] - pd.DateOffset(months=1)

# Round to nearest business day available in macro_df
# First make sure macro_df['Date'] is datetime and sorted
macro_df["Date"] = pd.to_datetime(macro_df["Date"])
macro_df = macro_df.sort_values("Date")

# Merge using the closest earlier or equal macro date (asof merge)
macro_df = macro_df.sort_values("Date")
real_df = real_df.sort_values("Macro_Merge_Date")
combined_df = pd.merge_asof(real_df, macro_df, left_on="Macro_Merge_Date", right_on="Date", direction="backward")
combined_df.drop(columns=["Macro_Merge_Date", "Date"], inplace=True) # drop helper columns

# We keep the true end-of-period date for the forecast, which might be useful for temporal alignment, modeling lags, 
# visualizations, and potentially for grouped time-series CV.

print("Combined shape:", combined_df.shape)
combined_df.head()
```

```python
combined_df.to_csv("final_dataset.csv", index=False)
```
